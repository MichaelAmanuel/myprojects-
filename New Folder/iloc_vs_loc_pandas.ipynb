**To complete the task, create a dictionary called data where the values should be pandas Series. Store the pandas Series in variables named names, age, and working_experience_yrs.**

```python
import pandas as pd


names = pd.Series(["Amy White", "Jack Stewart", "Richard Lauderdale", "Sara Johnson"])
age = pd.Series([50, 53, 35, 43])
working_experience_yrs = pd.Series([5, 8, 3, 10])

data = {
    "Name": names,
    "Age": age,
    "Working Experience (Yrs.)": working_experience_yrs
}

df = pd.DataFrame(data)


print(df)
```

```python
import pandas as pd

names = pd.Series(["Amy White", "Jack Stewart", "Richard Lauderdale", "Sara Johnson"])
age = pd.Series([50, 53, 35, 43])
working_experience_yrs = pd.Series([5, 8, 3, 10])


data = {
    "Name": names,
    "Age": age,
    "Working Experience (Yrs.)": working_experience_yrs
}


df = pd.DataFrame(data)

print(df)
```
