```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Wedge
from matplotlib.textpath import TextPath
from matplotlib.font_manager import FontProperties
from shapely.geometry import Polygon

# Define the color scheme
primary_color = '#1E90FF'  # Dodger blue for trust and professionalism
secondary_color = '#000000'  # Black for authority and strength
highlight_color = '#FF4500'  # Orange-red for alertness

# Create the figure and axis
fig, ax = plt.subplots()

# Draw a clock to represent 24/7 service
clock_outer = Circle((0.5, 0.6), 0.3, color=primary_color, fill=True)
ax.add_patch(clock_outer)
clock_inner = Circle((0.5, 0.6), 0.28, color='white', fill=True)
ax.add_patch(clock_inner)

# Draw clock hands
hour_hand = Wedge((0.5, 0.6), 0.2, 0, 90, width=0.05, color=secondary_color)
minute_hand = Wedge((0.5, 0.6), 0.25, 0, 180, width=0.03, color=secondary_color)
ax.add_patch(hour_hand)
ax.add_patch(minute_hand)

# Add the company name
font = FontProperties(family='sans-serif', weight='bold', size=15)
text_path = TextPath((0.05, 0), "24 Secure Solutions", prop=font)
text_patch = Polygon(text_path.vertices, closed=True, color=secondary_color)
ax.add_patch(text_patch)

# Customize the plot
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.3, 1.1)
ax.axis('off')

# Show the logo
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
